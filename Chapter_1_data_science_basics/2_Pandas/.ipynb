```python
df_gender = pd.get_dummies(df['gender']) # convert categorical col into numerical data
```

```python
# Create an example dataframe
data = {'group': ['A','A','A','A','A','A','B','B','B','B','B','C','C','C','C','C'],
       'marks': [1,4,5,7,5,5,6,1,4,5,6,7,4,6,4,6]}
df = pd.DataFrame(data)
df
```

```python
df.groupby('group')['marks'].min() # grouping between marks and groups
```

```python
def max_difference(col_data):
    return max(col_data)-min(col_data)
df.groupby('group')['marks'].apply(max_difference).plot(kind='bar')
```
