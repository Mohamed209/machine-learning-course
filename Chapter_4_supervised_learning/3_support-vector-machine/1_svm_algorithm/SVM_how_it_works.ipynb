## Kernel Trick

```python
# assume we have a dataset with x1,x2 features as in figure below

import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets.samples_generator import make_circles
X, y = make_circles(100, factor=.1, noise=.1)
#2D scatter plot
fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot()
ax.scatter(X[:,0], X[:,1], c = y, cmap = 'viridis')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_title('Original dataset')
```

`using svm as linear separator will not work , so here comes the concept of kernel function
and simply kernel is something like mapping function , that maps data from n space to m space (for example from 2D to 3D)`

`let us talk about very simple and eficient kernel is polynomial kernel`

$$\begin{eqnarray}
\phi(x_1,x_2) = (z_1,z_2,z_3) = (x_1,x_2,x_1^2+x_2^2)
\end{eqnarray}$$

```python
def kernel(X):
    return np.asarray((X[:,0], X[:,1], X[:,0]**2 + X[:,1]**2)).T

Z = kernel(X)

#3D scatter plot
fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(projection='3d')
ax.scatter3D(Z[:,0],Z[:,1], Z[:,2],c = y, cmap = 'viridis' )
ax.set_xlabel('$z_1$')
ax.set_ylabel('$z_2$')
ax.set_zlabel('$z_3$')
ax.set_title('Transformed dataset')
```
