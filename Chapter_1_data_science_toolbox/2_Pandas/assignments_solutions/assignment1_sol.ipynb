```python
import pandas as pd
```

```python
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, 10.5, 9, 20, 14.5, 7, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1]}

df=pd.DataFrame(exam_data)

df
```

```python
# print rows from 3:7 from columns attempts and score
df[['attempts','score']][3:7]
```

```python
# set the value of score column at row 4 to 15
df['score'].iloc[4]=15
df
```

```python
# add new column named "pass/fail" where its cells contains pass if score > 10 else cell contain fail
def pass_fail(cell):
    if cell > 10:
        return 'pass'
    else :
        return 'fail'
df['pass/fail']=df['score'].apply(lambda x : pass_fail(x))
df
```

```python
# if max score from 20 , create new column called "score_percentage" its cells contain perentage of the student score
df['percentage %']=df['score'].apply(lambda x : (x/20)*100)
df
```

```python
# make bar plot of student score on y axis and student name on x axis
df.plot.bar(y='score',x='name')
```

```python
# delete column "attempts" from dataframe
df.drop(['attempts'],inplace=True,axis=1)
df
```

```python
'''7. divide the dataframe into to equal parts first part has all columns and rows from start until middle of the dataframe and second part contains all columns and rows from middle to the end of the dataframe'''
midpoint=len(df)//2 # middle index of the dataframe
part1=df[0:midpoint]
print(part1)
part2=df[midpoint:]
print(part2)
   
```
